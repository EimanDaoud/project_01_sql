```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### **1st Chart:** Average Salary Distribution for Top 10 Paying Data Analyst Jobs in 2023

```python
df1 = pd.read_csv('Top-paying Data Analyst roles.csv')
```

```python
# Create the plot
plt.figure(figsize=(8, 5))

# Apply a color map to the bars (e.g., Blues)
sns.barplot(y=df1['job_title'],
            x=df1['salary_year_avg'], 
            hue=df1['salary_year_avg'], 
            palette='viridis_d', 
            width=0.55
            )

# Customize the title
plt.title(
    'Average Salary Distribution for Top 10 Paying Data Analyst Jobs in 2023', 
    fontsize=12, 
    color='darkslategrey', 
    fontweight='bold', 
    loc='center',
    family = 'monospace',
    pad=13
)

# Invert y-axis for top skills to appear at the top
plt.gca().invert_yaxis()

# Customize font size, color, and weight for axis tick labels
plt.xticks(fontsize=12, color='darkblue', fontweight='bold')
plt.yticks(fontsize=10, color='grey', fontweight='normal')

# Set grid and place it behind the bars
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().set_axisbelow(True)

# Customize the axis labels
plt.xlabel('Average Yearly Salary ($)',fontsize=11, color='grey', fontweight='ultralight')
plt.ylabel('')


# Invert y-axis for better readability
plt.gca().invert_yaxis()

# Customize font size, color, and weight for axis tick labels
plt.xticks(fontsize=10, color='grey', fontweight='normal')
plt.yticks(fontsize=10, color='grey', fontweight='ultralight', family='sans')

# Remove the color legend
plt.legend([], frameon=False)

plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)

plt.gca().tick_params(axis='x', which='both', length=0)
plt.gca().tick_params(axis='y', which='both', length=0)
# Show the plot
plt.tight_layout()
plt.show()
```

#### **2nd Chart:** Skill Count for Top 10 Paying Data Analyst Jobs in 2023

```python
df2 = pd.read_csv('Skills for Top Paying Jobs.csv')
```

```python
skills_count = df2.skills.value_counts().reset_index().head(10)
skills_count['SKILLS_U'] = skills_count['skills'].str.upper()
```

```python
skills_count.drop('skills', axis=1)
```

```python
# Create the plot
plt.figure(figsize=(8, 4))

# Apply a color map to the bars (e.g., Blues)
sns.barplot(y=skills_count['SKILLS_U'],
            x=skills_count['count'], 
            hue=skills_count['count'], 
            palette='viridis_d', 
            width=0.55
            )

# Customize the title
plt.title(
    'Skill Count for Top 10 Paying Data Analyst Jobs in 2023', 
    fontsize=12, 
    color='darkslategrey', 
    fontweight='bold', 
    loc='center',
    family = 'monospace'
)

# Invert y-axis for top skills to appear at the top
plt.gca().invert_yaxis()

# Customize font size, color, and weight for axis tick labels
plt.xticks(fontsize=12, color='darkblue', fontweight='bold')
plt.yticks(fontsize=10, color='grey', fontweight='normal')

# Set grid and place it behind the bars
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().set_axisbelow(True)

# Customize the axis labels
#plt.xlabel('Frequency',fontsize=14, color='midnightblue', fontweight='normal')
plt.xlabel('')
plt.ylabel('')

# Invert y-axis for better readability
plt.gca().invert_yaxis()

# Customize font size, color, and weight for axis tick labels
plt.xticks(fontsize=10, color='grey', fontweight='normal')
plt.yticks(fontsize=10, color='grey', fontweight='ultralight', family='sans')

# Remove the color legend
plt.legend([], frameon=False)

plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)

plt.gca().tick_params(axis='x', which='both', length=0)
plt.gca().tick_params(axis='y', which='both', length=0)
# Show the plot
plt.tight_layout()
plt.show()
```
